```python
from ultralytics import YOLO
from pathlib import Path
import shutil

SAVE_DIR = "/content/drive/MyDrive/AgriVision/object_detection_result"

model = YOLO("yolov8n.pt")

results = model.train(
    data="/content/dataset/PlantVillage_for_object_detection/SplitDataset/data.yaml",
    epochs=20,
    imgsz=640,
    batch=16,
    device=0,
    workers=2,
    project=SAVE_DIR,
    name="yolo",
    save=True
)

best = Path(results.save_dir) / "weights" / "best.pt"
final = Path(SAVE_DIR) / "yolo.pt"
shutil.copy2(best, final)

print("✅ Final model saved at:", final)
```
